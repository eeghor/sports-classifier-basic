```python
import pandas as pd
from nltk.corpus import stopwords
d = pd.read_csv("data/golf_events_31082017.csv", dtype=str).fillna("").applymap(lambda _: _.lower())
abbrs = {"adm": "admission", "int": "international", "int'l": "international", "champ": "championship", "ent": "entertainment"}
```

```python
d.head()
```

```python
stop_words = set(stopwords.words('english'))

def deabbrevi8(s):
    return " ".join([w if w not in abbrs else abbrs[w] for w in s.lower().split()])

for i, row in enumerate(d.iterrows()):
    res = " ".join(row[1][1].split()) # starting from 'primary_show_desc'
    for j, col in enumerate(d.columns[2:]):  # start from 'title_who'
        if not (" ".join(row[1][col].split()) in res):
            res = " ".join([res, row[1][col]]) 
    res = " ".join([w for w in res.split() if w.isalpha() and (w not in stop_words)])
    print(res)             
    if i == 5:
        break
```
